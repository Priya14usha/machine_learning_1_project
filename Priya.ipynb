```python
import pandas as pd
import numpy as np
import sklearn
```

```python
# calling a dataset from sklearn
from sklearn.datasets import load_boston
df=load_boston()
df
```

```python
df.keys()  #displays/returns the keys present in boston dataset dictionay
```

```python
df['DESCR']  #info about particular key
```

```python
df['target']
```

```python
# putting data in tabular form
a=pd.DataFrame(df.data,columns=df.feature_names)
a.head()
```

```python
#adding a new coloumn for target values as taret is not present in data
a["Price"] =df.target
a.head()
```

```python
a.isnull()  #checking missing values
```

```python
a.isnull().sum()   
```

```python
#calling a module for splitting the training and testing data
from sklearn.model_selection import train_test_split
x=a.drop('Price',axis=1)
y=a['Price']
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.16,random_state=5)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#calling linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
model=LinearRegression()
```

```python
#fitting the training data into the model
c1=model.fit(x_train,y_train)
c1
```

```python
y_train_predict=model.predict(x_train)   #predictions on training data
rmse=np.sqrt(mean_squared_error(y_train,y_train_predict))

print("The model performance on training data")
print("RMSE ={}".format(rmse))
```

```python
y_test_predict=model.predict(x_test)   #predictions on testing data
rmse=np.sqrt(mean_squared_error(y_test,y_test_predict))

print("The model performance on testing data")
print("RMSE ={}".format(rmse))
```
